```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical

# Исходная строка
text = "hellohellohello"

# Создаем словарь символов и их индексов
chars = sorted(list(set(text)))  # Получаем уникальные символы
char_to_idx = {c: i for i, c in enumerate(chars)}  # Создаем словарь символ -> индекс
idx_to_char = {i: c for i, c in enumerate(chars)}  # Создаем словарь индекс -> символ

# Преобразуем строку в последовательности индексов
input_seq = [char_to_idx[c] for c in text]

# Создаем обучающие примеры (последовательности и следующий символ)
X = []
y = []

seq_length = 4  # Длина последовательности
for i in range(len(input_seq) - seq_length):
    X.append(input_seq[i:i+seq_length])  # Последовательность длиной 4 символа
    y.append(input_seq[i+seq_length])    # Следующий символ

# Преобразуем в numpy массивы
X = np.array(X)
y = np.array(y)

# Преобразуем y в формат one-hot encoding для выхода сети
y = to_categorical(y, num_classes=len(chars))

# Изменяем форму X для подачи в RNN (samples, time_steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))  # 1 фича на timestep (индекс символа)

# Создаем простую модель RNN
model = Sequential()
model.add(SimpleRNN(32, input_shape=(seq_length, 1)))  # Один RNN слой
model.add(Dense(len(chars), activation='softmax'))     # Выходной слой с количеством классов, равным числу символов

# Компиляция модели
model.compile(loss='categorical_crossentropy', optimizer='adam')

# Обучение модели
model.fit(X, y, epochs=100, verbose=1)

# Функция для предсказания следующего символа
def predict_next_char(sequence):
    x = np.array([char_to_idx[c] for c in sequence])
    x = np.reshape(x, (1, seq_length, 1))  # Преобразуем для подачи в модель
    prediction = model.predict(x, verbose=0)
    idx = np.argmax(prediction)  # Получаем индекс предсказанного символа
    return idx_to_char[idx]  # Возвращаем предсказанный символ

# Тестируем на примерах
test_sequence = "hell"
print(f"Последовательность: {test_sequence}")
print(f"Предсказанный следующий символ: {predict_next_char(test_sequence)}")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

text = """
In a small village in the mountains lived an old man named John. John had many stories to tell.
He knew the secrets of the forest and often spoke to the animals. One day, a young boy named Tom came to visit John.
Tom was curious about the legends of the mountains and asked John about the secret treasure hidden in the caves.
John smiled and said, 'The treasure is not what you think it is, my boy.'
"""

# Токенизация текста
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

# Преобразование текста в последовательность индексов
sequence_data = tokenizer.texts_to_sequences([text])[0]

# Подготовка обучающих данных: последовательности слов (например, 5 слов → предсказываем 6-е слово)
sequence_length = 5
X = []
y = []

for i in range(sequence_length, len(sequence_data)):
    X.append(sequence_data[i-sequence_length:i])
    y.append(sequence_data[i])

X = np.array(X)
y = np.array(y)

# Параметры модели
vocab_size = len(tokenizer.word_index) + 5  # Размер словаря (количество уникальных слов)
embedding_dim = 100  # Размерность векторного представления слов

# Создаем модель LSTM
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequence_length))
model.add(LSTM(256, return_sequences=False))  # Слой LSTM
model.add(Dense(vocab_size, activation='softmax'))  # Выходной слой для предсказания следующего слова

# Компиляция модели
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Обучение модели
model.fit(X, y, epochs=200, verbose=2)

# Функция для генерации текста (ответов на вопросы)
def generate_text(model, tokenizer, seed_text, next_words=10):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=sequence_length, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)
        output_word = tokenizer.index_word[predicted[0]]
        seed_text += " " + output_word
    return seed_text

# Задаем вопрос (начало текста)
question = "Tom asked John"
answer = generate_text(model, tokenizer, question, next_words=10)

print("Вопрос:", question)
print("Ответ:", answer)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.metrics import Precision, Recall
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# Генерация данных (временной ряд синусоидальной функции)
def generate_data():
    x = np.linspace(0, 100, 1000)
    y = np.sin(x)
    return x, y

# Подготовка данных
x, y = generate_data()
scaler = MinMaxScaler(feature_range=(0, 1))
y_scaled = scaler.fit_transform(y.reshape(-1, 1))

# Преобразование для классификации
threshold = 0.5  # Порог для разделения классов
y_class = (y_scaled > threshold).astype(int)

# Создание окон временного ряда
def create_dataset(data, time_step=10):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

# Параметры окна
time_step = 50
X_class, Y_class = create_dataset(y_class, time_step)
X_class = np.reshape(X_class, (X_class.shape[0], X_class.shape[1], 1))

# Разделение на обучающую и тестовую выборки
X_train, X_test, Y_train, Y_test = train_test_split(X_class, Y_class, test_size=0.2, random_state=42)

# Функция компиляции модели
def compile_model(model):
    model.compile(optimizer='adam', loss='binary_crossentropy', 
                  metrics=['accuracy', Precision(), Recall()])
    return model

# Функция для оценки модели
def evaluate_model(model, X_train, Y_train, X_test, Y_test, use_early_stopping=False):
    # Early stopping
    callbacks = [EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)] if use_early_stopping else None

    # Обучение модели
    history = model.fit(X_train, Y_train, epochs=20, batch_size=32, 
                        validation_data=(X_test, Y_test), verbose=0, callbacks=callbacks)
    
    # Получение метрик на обучающем наборе
    train_eval = model.evaluate(X_train, Y_train, verbose=0)
    train_loss, train_acc, train_precision, train_recall = train_eval
    train_f1 = f1_score(Y_train, model.predict(X_train).round())
    
    # Получение метрик на тестовом наборе
    test_eval = model.evaluate(X_test, Y_test, verbose=0)
    test_loss, test_acc, test_precision, test_recall = test_eval
    test_f1 = f1_score(Y_test, model.predict(X_test).round())

    results = {
        "Train Loss": train_loss, "Train Accuracy": train_acc, 
        "Train Precision": train_precision, "Train Recall": train_recall, "Train F1": train_f1,
        "Test Loss": test_loss, "Test Accuracy": test_acc, 
        "Test Precision": test_precision, "Test Recall": test_recall, "Test F1": test_f1
    }
    return results, history

# Создание и компиляция моделей
rnn_model = Sequential([SimpleRNN(50, return_sequences=True, input_shape=(time_step, 1)), 
                        Dropout(0.2), SimpleRNN(50), Dropout(0.2), Dense(1, activation='sigmoid')])
gru_model = Sequential([GRU(50, return_sequences=True, input_shape=(time_step, 1)), 
                        Dropout(0.2), GRU(50), Dropout(0.2), Dense(1, activation='sigmoid')])
lstm_model = Sequential([LSTM(50, return_sequences=True, input_shape=(time_step, 1)), 
                         Dropout(0.2), LSTM(50), Dropout(0.2), Dense(1, activation='sigmoid')])
two_level_lstm_model = Sequential([LSTM(50, return_sequences=True, input_shape=(time_step, 1)),
                                   Dropout(0.2), LSTM(50, return_sequences=True),
                                   Dropout(0.2), LSTM(50), Dropout(0.2), Dense(1, activation='sigmoid')])
early_stopping_lstm_model = Sequential([LSTM(50, return_sequences=True, input_shape=(time_step, 1)), 
                                        Dropout(0.2), LSTM(50), Dropout(0.2), Dense(1, activation='sigmoid')])

# Компиляция всех моделей
rnn_model = compile_model(rnn_model)
gru_model = compile_model(gru_model)
lstm_model = compile_model(lstm_model)
two_level_lstm_model = compile_model(two_level_lstm_model)
early_stopping_lstm_model = compile_model(early_stopping_lstm_model)

# Оценка моделей
rnn_results, rnn_history = evaluate_model(rnn_model, X_train, Y_train, X_test, Y_test)
gru_results, gru_history = evaluate_model(gru_model, X_train, Y_train, X_test, Y_test)
lstm_results, lstm_history = evaluate_model(lstm_model, X_train, Y_train, X_test, Y_test)
two_level_lstm_results, two_level_lstm_history = evaluate_model(two_level_lstm_model, X_train, Y_train, X_test, Y_test)
early_stopping_lstm_results, early_stopping_lstm_history = evaluate_model(early_stopping_lstm_model, X_train, Y_train, X_test, Y_test, use_early_stopping=True)

# Вывод результатов
results_df = pd.DataFrame({
    "SimpleRNN": rnn_results,
    "GRU": gru_results,
    "LSTM": lstm_results,
    "Two-Level LSTM": two_level_lstm_results,
    "LSTM w/ Early Stopping": early_stopping_lstm_results
})
print(results_df)

# Визуализация изменения ошибки (Loss) и точности (Accuracy)
plt.figure(figsize=(14, 7))

# График для метрики Loss
plt.subplot(1, 2, 1)
plt.plot(rnn_history.history['loss'], label="SimpleRNN Train Loss")
plt.plot(rnn_history.history['val_loss'], label="SimpleRNN Test Loss")
plt.plot(gru_history.history['loss'], label="GRU Train Loss")
plt.plot(gru_history.history['val_loss'], label="GRU Test Loss")
plt.plot(lstm_history.history['loss'], label="LSTM Train Loss")
plt.plot(lstm_history.history['val_loss'], label="LSTM Test Loss")
plt.plot(two_level_lstm_history.history['loss'], label="Two-Level LSTM Train Loss")
plt.plot(two_level_lstm_history.history['val_loss'], label="Two-Level LSTM Test Loss")
plt.plot(early_stopping_lstm_history.history['loss'], label="LSTM w/ Early Stopping Train Loss")
plt.plot(early_stopping_lstm_history.history['val_loss'], label="LSTM w/ Early Stopping Test Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Изменение значения Loss")
plt.legend()

# График для метрики Accuracy
plt.subplot(1, 2, 2)
plt.plot(rnn_history.history['accuracy'], label="SimpleRNN Train Accuracy")
plt.plot(rnn_history.history['val_accuracy'], label="SimpleRNN Test Accuracy")
plt.plot(gru_history.history['accuracy'], label="GRU Train Accuracy")
plt.plot(gru_history.history['val_accuracy'], label="GRU Test Accuracy")
plt.plot(lstm_history.history['accuracy'], label="LSTM Train Accuracy")
plt.plot(lstm_history.history['val_accuracy'], label="LSTM Test Accuracy")
plt.plot(two_level_lstm_history.history['accuracy'], label="Two-Level LSTM Train Accuracy")
plt.plot(two_level_lstm_history.history['val_accuracy'], label="Two-Level LSTM Test Accuracy")
plt.plot(early_stopping_lstm_history.history['accuracy'], label="LSTM w/ Early Stopping Train Accuracy")
plt.plot(early_stopping_lstm_history.history['val_accuracy'], label="LSTM w/ Early Stopping Test Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Изменение значения Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping

text = """
In a small village in the mountains lived an old man named John. John had many stories to tell.
He knew the secrets of the forest and often spoke to the animals. One day, a young boy named Tom came to visit John.
Tom was curious about the legends of the mountains and asked John about the secret treasure hidden in the caves.
John smiled and said, 'The treasure is not what you think it is, my boy.'
"""

# Токенизация текста
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

# Преобразование текста в последовательность индексов
sequence_data = tokenizer.texts_to_sequences([text])[0]

# Подготовка обучающих данных: последовательности слов (например, 5 слов → предсказываем 6-е слово)
sequence_length = 5
X = []
y = []

for i in range(sequence_length, len(sequence_data)):
    X.append(sequence_data[i-sequence_length:i])
    y.append(sequence_data[i])

X = np.array(X)
y = np.array(y)

# Параметры модели
vocab_size = len(tokenizer.word_index) + 5  # Размер словаря (количество уникальных слов)
embedding_dim = 100  # Размерность векторного представления слов

# Создаем модель с двухуровневой LSTM
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequence_length))
model.add(LSTM(256, return_sequences=True))  # Первый слой LSTM с возвратом всех последовательностей
model.add(Dropout(0.2))  # Dropout для регуляризации
model.add(LSTM(256, return_sequences=False))  # Второй слой LSTM
model.add(Dropout(0.2))
model.add(Dense(vocab_size, activation='softmax'))  # Выходной слой для предсказания следующего слова

# Компиляция модели
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Ранняя остановка
early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)

# Обучение модели
model.fit(X, y, epochs=200, verbose=2, callbacks=[early_stopping])

# Функция для генерации текста (ответов на вопросы)
def generate_text(model, tokenizer, seed_text, next_words=10):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=sequence_length, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)
        output_word = tokenizer.index_word.get(predicted[0], '')
        seed_text += " " + output_word
    return seed_text

# Задаем вопрос (начало текста)
question = "Tom asked John"
answer = generate_text(model, tokenizer, question, next_words=10)

print("Вопрос:", question)
print("Ответ:", answer)
```
