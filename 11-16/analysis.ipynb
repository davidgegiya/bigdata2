```python
# Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Загрузка данных из CSV файла
data = pd.read_csv('mammoth.csv')  # Укажите путь к вашему CSV файлу

# Извлечение признаков для анализа
X = data[['x', 'y', 'z']]

# 2. Описание данных
print("Информация о данных:")
print(X.info())  # Проверяем типы данных и наличие пропущенных значений
print("\nОсновная статистика данных:")
print(X.describe())  # Описательная статистика для числовых данных
print("\nКоличество пропущенных значений:")
print(X.isnull().sum())  # Проверка на пропуски

# 3. Визуализация распределения данных

# Гистограмма для каждого признака
X.hist(figsize=(10, 6), bins=20)
plt.suptitle('Распределение признаков x, y, z')
plt.show()

# 4. Корреляция и зависимость

# Корреляционная матрица для признаков x, y, z
corr_matrix = X.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Корреляционная матрица для признаков x, y, z')
plt.show()

# Парные графики для x, y, z
sns.pairplot(X)
plt.suptitle('Парные графики для признаков x, y, z', y=1.02)
plt.show()

# 5. Boxplot для анализа выбросов
plt.figure(figsize=(10, 6))
sns.boxplot(data=X)
plt.title('Boxplot для признаков x, y, z')
plt.show()
```
