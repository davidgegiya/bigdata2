```python
from sklearn.manifold import TSNE
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo 
from sklearn.preprocessing import OneHotEncoder
  
# DATESET_ID = 82
DATESET_ID = 90

post_operative_patient = fetch_ucirepo(id=DATESET_ID) 
  
X = post_operative_patient.data.features 
y = post_operative_patient.data.targets

# Преобразование категориальных признаков в числовые с помощью One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False)
X_encoded = encoder.fit_transform(X)
# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

settings = [
    {
        "perplexity": 30,
        "early_exaggeration": 12,
        "learning_rate": 200,
        "max_iter": 1000
    },
    {
        "perplexity": 50,
        "early_exaggeration": 15,
        "learning_rate": 200,
        "max_iter": 5000
    },
    {
        "perplexity": 5,
        "early_exaggeration": 8,
        "learning_rate": 200,
        "max_iter": 1000
    },
    {
        "perplexity": 40,
        "early_exaggeration": 12,
        "learning_rate": 400,
        "max_iter": 5000
    },
    {
        "perplexity": 70,
        "early_exaggeration": 8,
        "learning_rate": 300,
        "max_iter": 3000
    },
    {
        "perplexity": 30,
        "early_exaggeration": 12,
        "learning_rate": 200,
        "max_iter": 500
    },
]

# Создаем фигуру для 6 подграфиков
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Применяем t-SNE с различными настройками и строим графики
for i, ax in enumerate(axes.flat):
    tsne = TSNE(n_components=2, random_state=42, **settings[i])
    X_tsne = tsne.fit_transform(X_scaled)
    
    # Отображаем результат на подграфике
    scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], c=pd.factorize(y.squeeze())[0], cmap='tab10', alpha=0.7)
    ax.set_title(f"t-SNE perplexity={settings[i]['perplexity']}, lr={settings[i]['learning_rate']}")
    
# plt.colorbar(scatter, ax=axes.ravel().tolist(), orientation='horizontal', fraction=0.05, pad=0.08)

# Отображаем графики
plt.tight_layout()
plt.show()
```
